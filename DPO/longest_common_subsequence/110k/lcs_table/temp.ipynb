```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import os

# 定义所有维度
dimensions = [
    'robustness',
    'standardization',
    'simplicity',
    'modularity',
    'functionality',
    'efficiency',
    'comment'
]

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 创建输出目录
output_dir = 'lcs_score_plots'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 处理每个维度
for dimension in dimensions:
    # 读取JSONL文件
    lcs_scores = []
    file_path = f'{dimension}_lcs_results.jsonl'
    
    with open(file_path, 'r') as f:
        for line in f:
            data = json.loads(line)
            scores = data['lcs_score']
            for score in scores:
                if 0 <= score <= 30:  # 修改为0-30分范围
                    lcs_scores.append(score)

    # 计算总数
    total_scores = len(lcs_scores)
    print(f"{dimension}维度 0-30分的总数: {total_scores}")

    # 创建直方图
    plt.figure(figsize=(12, 6))
    sns.histplot(lcs_scores, bins=31, kde=True)
    plt.title(f'{dimension}维度 LCS分数分布直方图')
    plt.xlabel('LCS分数')
    plt.ylabel('频数')
    plt.grid(True, alpha=0.3)

    # 设置横轴刻度，显示所有整数
    plt.xticks(range(0, 31, 1))
    
    # 调整布局
    plt.tight_layout()
    
    # 保存图片
    plt.savefig(f'{output_dir}/{dimension}_lcs_score_histogram.png')
    plt.close()

print("所有维度的图表已生成完成！")
```
