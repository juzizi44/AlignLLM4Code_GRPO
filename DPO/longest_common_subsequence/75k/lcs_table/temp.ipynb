```python
import json
import matplotlib.pyplot as plt
import seaborn as sns
import os
import math

# 定义所有维度
dimensions = [
    'robustness',
    'standardization',
    'simplicity',
    'modularity',
    'functionality',
    'efficiency',
    'comment'
]

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 创建输出目录
output_dir = 'lcs_score_plots'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 创建子图画布：每行一个，共7个
fig, axes = plt.subplots(len(dimensions), 1, figsize=(12, 4 * len(dimensions)))
if len(dimensions) == 1:
    axes = [axes]  # 保证是可迭代的

# 遍历每个维度并绘图
for idx, dimension in enumerate(dimensions):
    lcs_scores = []
    file_path = f'{dimension}_lcs_results.jsonl'

    with open(file_path, 'r') as f:
        for line in f:
            data = json.loads(line)
            scores = data['lcs_score']
            for score in scores:
                rounded = round(score)
                if 0 <= rounded <= 30:
                    lcs_scores.append(rounded)

    total_scores = len(lcs_scores)
    print(f"{dimension}维度 0-30分的总数: {total_scores}")

    ax = axes[idx]
    sns.histplot(lcs_scores, bins=range(0, 32), kde=True, ax=ax)
    ax.set_title(f'{dimension}维度 LCS分数分布直方图')
    ax.set_xlabel('LCS分数')
    ax.set_ylabel('频数')
    ax.set_xticks(range(0, 31, 1))
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig(f'{output_dir}/all_dimensions_lcs_score_histograms.png')
plt.close()

print("所有维度的合并图表已生成完成！")
```
